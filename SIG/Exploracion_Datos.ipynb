# Leemos el Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_train.csv')
df.head()
```

# Análisis de variables

* 1 y 4 **SOJA** (temprana, tardía)
* 2 y 3 **MAIZ** (temprano, tardío)
* 5 **GIRASOL**
* Ignorar las clases 10 y 20

```python
# Analisamos la cantidad de registros por tipo de cultivo
df['cultivo'].value_counts().plot(kind='bar')
```

```python
df.describe()
```

# Analizamos Bandas 4 y 8 por cultivo
**Banda 4**: Rojo

**Banda 8**: Infrarojo cercano (NIR)

```python
# Variables
labels = ['B4-Oct', 'B8-Oct', 'B4-Nov', 'B8-Nov', 
          'B4-Dic', 'B8-Dic', 'B4-Ene', 'B8-Ene',
          'B4-Feb', 'B8-Feb', 'B4-Mar', 'B8-Mar' ]
colors = ['red', 'tomato'] * 6
```

## Cultivo de Soja Temprano

```python
soja_temp = df[(df['cultivo'] == 1)].filter(regex='B8_|B4_')
```

```python
fig = plt.figure(figsize =(14, 7)) 
fig.suptitle('Soja Temprano', fontsize=16, fontweight='bold')

data = [soja_temp['B4_2020-10-01'], soja_temp['B8_2020-10-01'],
        soja_temp['B4_2020-11-01'], soja_temp['B8_2020-11-01'],
        soja_temp['B4_2020-12-01'], soja_temp['B8_2020-12-01'],
        soja_temp['B4_2021-01-01'], soja_temp['B8_2021-01-01'],
        soja_temp['B4_2021-02-20'], soja_temp['B8_2021-02-20'],
        soja_temp['B4_2021-03-17'], soja_temp['B8_2021-03-17']]

box = plt.boxplot(data, labels=labels, patch_artist=True)
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
soja_temp.describe()
```
